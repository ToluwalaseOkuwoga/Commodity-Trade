```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
```

```python
df = pd.read_csv("commodity_trade_statistics_data.csv")
```

```python
df['weight_kg'].fillna(df['weight_kg'].mean(), inplace=True)
```

```python
df['quantity'].fillna(df['quantity'].mean(), inplace=True)
```

```python
df[df['country_or_area']=='EU-28']
```

```python
df = df[df.country_or_area != 'EU-28']
```

```python
df.shape
```

```python
# Renaming the country column
df.rename({'country_or_area':'country'},axis =1, inplace=True)

# Replacing all China areas with China
df.country.replace({'China, Hong Kong SAR':'China'},inplace=True)

df.country.replace({'Rep. of Korea':'South Korea'},inplace=True)
df.country.replace({'Russian Federation':'Russia'},inplace=True)
```

```python
df_country=df.groupby(['country'],as_index=False)['weight_kg','trade_usd'].agg('sum')
```

```python
df_country=df_country.sort_values(['trade_usd'],ascending=False)
```
